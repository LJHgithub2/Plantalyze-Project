```python
from ultralytics import YOLO

# 학습된 가중치 파일 적용
model = YOLO('/content/drive/MyDrive/Colab Notebooks/Project/Chungbuk University/Capstone Design/AI_Model/PlantDiseaseDetection/models/leaf_detection.pt')
```

```python
import glob
import os
def get_img_path_list(root_dir, extensions):

  img_list = list()
  for ext_label in extensions:
      img_list.extend(glob.glob(os.path.join(root_dir, f"*.{ext_label}")))

  return img_list
```

```python
# 해당 폴더 안의 모든 이미지 파일의 잎 위치 탐지
test_img_path="/content/drive/MyDrive/Colab Notebooks/Project/Chungbuk University/Capstone Design/AI_Model/PlantDiseaseDetection/data/test_img/"

extensions = ["json", "txt", "jpg", "JPG", "png", "PNG"]

image_paths = get_img_path_list(test_img_path, extensions)
print(image_paths)
```

```python
import cv2
from google.colab.patches import cv2_imshow

def plot_images(model, image_paths):
    num_images = len(image_paths)

    for i, image_path in enumerate(image_paths, 1):
        # 이미지에 대한 객체 검출 결과 얻기
        results = model(image_path)
        plots = results[0].plot()
        cv2_imshow(plots)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# 함수 호출하여 이미지 플롯 표시
plot_images(model, image_paths)
```
